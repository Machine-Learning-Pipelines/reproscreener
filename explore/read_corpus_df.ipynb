```python
from pathlib import Path
import pandas as pd
from reproscreener.gold_standard import get_gold_standard_ids_from_manual
```

```python
dtypes_scrape = {'id': str}
path_gold_standard = Path("../case-studies/arxiv-corpus/gold_standard/")

path_manual_eval = Path("../case-studies/arxiv-corpus/manual_eval.csv")
# manual_ids = get_gold_standard_ids_from_manual(path_manual_eval)
# eval = pd.read_csv(path_corpus_and / 'output/repro_eval_tex.csv')[['id', 'link_count', 'found_links']]
manual_eval = pd.read_csv(path_manual_eval)
scrape_df = pd.read_csv(path_gold_standard / 'scrape_df.csv', dtype=dtypes_scrape)
```

```python
# drop the 0th row, which is the column names
manual_eval = manual_eval.drop(0)

dtypes_manual = {'paper': str, 'article_link_avail': float, 'pwc_link_avail': float, 'pwc_link_match': float, 'result_replication_code_avail': float}
manual_eval = manual_eval.astype(dtypes_manual)

# Convert the 'paper' and 'id' columns into sets
manual_set = set(manual_eval['paper'])
scrape_set = set(scrape_df['id'])

# Find the intersection (i.e., matches)
matches = manual_set.intersection(scrape_set)

# Find the non-matches
non_matches_manual = manual_set.difference(scrape_set)
non_matches_scrape = scrape_set.difference(manual_set)

# Convert matches and non-matches into dataframes for visual presentation
matches_df = pd.DataFrame(sorted(list(matches)), columns=['matches'])
non_matches_manual_df = pd.DataFrame(sorted(list(non_matches_manual)), columns=['non_matched_manual'])
non_matches_scrape_df = pd.DataFrame(sorted(list(non_matches_scrape)), columns=['non_matched_scrape'])

matches_df
```
